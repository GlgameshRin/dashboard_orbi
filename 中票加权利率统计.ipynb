## 读取csv文件
做分析一般用.csv格式作为数据源而不用.xlsx格式,因为excel会有格式转换的问题

```python
df1 = pd.read_csv("中票加权利率统计0422.csv", encoding='gbk') 

# 中文的情况下需要用gbk编码，不然会乱码
```

```python
df1.head()

# 看看数据跟我们想的是否一致
```

```python
df2 = df1[df1['票面利率(%)'] >= 4.5] # 将利率 >= 4.92 的债券筛选出来
df2 = df2.reset_index(drop=True) # 正确的设置索引
df2 
```

```python
df2.describe() # pandas 内置的函数，可以看到各个列的基本分析，标准差std，平均值mean等
```

```python
df2['发行规模(亿)'].sum() # 发行规模的总和
```

## 对我们感兴趣的数据画图

```python
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题

plt.barh(df2["债券简称"],df2['发行规模(亿)']) # bar是竖着的柱状图 barh是横着的柱状图 plot是折线图
# plt.xticks(rotation=30)

plt.xlabel("发行规模 (亿)")
plt.ylabel("债券简称")
```

```python
X = df2["债券简称"]
Y = df2['发行规模(亿)']/df2['计划发行规模(亿)']

color_list = []
for i in Y:
    if i != 1:
        color = "red"
    else:
        color = "blue"
        
    color_list.append(color)


plt.barh(X,Y,color=color_list) # bar是竖着的柱状图 barh是横着的柱状图 plot是折线图
# plt.xticks(rotation=30)

plt.xlabel("发行规模/计划发行规模 (亿)")
plt.ylabel("债券简称")
```

```python
X = [a.replace("21","").replace("MTN001","") for a in df2["债券简称"]]

Y1 = df2['发行期限(年)']
Y2 = df2['票面利率(%)']


plt.bar(X,Y1) # bar是竖着的柱状图 barh是横着的柱状图 plot是折线图
plt.plot(X,Y2,color='red') # bar是竖着的柱状图 barh是横着的柱状图 plot是折线图

plt.xticks(rotation=30) # 将名称旋转30度不然塞不下

plt.legend(['票面利率(%)','发行期限(年)']) # 右上角legend

# 可以看到发行年限和票面利率之间的关系不明显
```
